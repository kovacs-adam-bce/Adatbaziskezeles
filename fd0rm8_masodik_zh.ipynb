1\. feladat



```sql
SELECT SZALLAS_ID, SZALLAS_NEV, CSILLAGOK_SZAMA,
IIF(CSILLAGOK_SZAMA>=3, 'Igen', 'Nem') AS 'Legalább 3 csillagos-e'
FROM Szallashely
WHERE SZALLAS_ID IN
(SELECT SZALLAS_ID
FROM Szoba
WHERE FEROHELY <>3
)
```

2\. feladat



```sql
SELECT SZOBA_ID, FOGLALAS_PK, DATEDIFF(DAY,METTOL,MEDDIG) AS 'Eltelt napok száma',
LAG(DATEDIFF(DAY,METTOL,MEDDIG),1,0) OVER (PARTITION BY SZOBA_ID ORDER BY FOGLALAS_PK) AS 'Előző foglaláskor eltelt napok száma'
FROM Szoba sz JOIN Foglalas f ON f.SZOBA_FK=sz.SZOBA_ID
```
